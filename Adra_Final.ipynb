```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

def split_into_chunks(extracted_information, max_tokens_per_chunk=264):
  """
  Splits a large string of text into chunks of sentences with a maximum token limit.

  Args:
      extracted_information: The long string of text to be chunked.
      max_tokens_per_chunk: The maximum number of tokens allowed in each chunk (default 480).

  Returns:
      A list of strings, where each string is a chunk of text not exceeding the limit.
  """

  sentences = sent_tokenize(extracted_information)
  chunks = []
  current_chunk = []
  chunk_num = 1

  for sentence in sentences:
    tokens = word_tokenize(sentence)

    if len(current_chunk) + len(tokens) <= max_tokens_per_chunk:
      current_chunk.extend(tokens)
    else:
      chunks.append(" ".join(current_chunk))
      current_chunk = tokens
      chunk_num += 1

  if current_chunk:
    chunks.append(" ".join(current_chunk))

  return chunks



chunks = split_into_chunks(extractedInformation)
num_chunks = len(chunks)

print("Number of chunks:", num_chunks)

for i, chunk in enumerate(chunks, start=1):
  tokens = word_tokenize(chunk)
  num_tokens_in_chunk = len(tokens)
  print(f"Chunk {i}:")
  print(f"\tNumber of tokens: {num_tokens_in_chunk}")
  print(f"\tContent: {chunk}")
  print()
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-large-uncased-whole-word-masking-finetuned-squad")
model = AutoModelForQuestionAnswering.from_pretrained("google-bert/bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def answer_question(question, context):

    inputs = tokenizer(question, context, return_tensors="pt")
    outputs = model(**inputs)
    start_logits = outputs.start_logits.squeeze().tolist()
    end_logits = outputs.end_logits.squeeze().tolist()

    best_start_index = max(range(len(start_logits)), key=start_logits.__getitem__)
    best_end_index = max(range(len(end_logits)), key=end_logits.__getitem__)

    if start_logits[best_start_index] < 0.1 or end_logits[best_end_index] < 0.1:
        return ""

    answer = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][best_start_index:best_end_index+1])
    answer = " ".join(answer)

    return answer

def get_most_similar_chunk(question, chunks):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(chunks + [question])
    similarity_scores = cosine_similarity(tfidf_matrix[:-1], tfidf_matrix[-1])

    most_similar_chunk_index = np.argmax(similarity_scores)
    highest_score = similarity_scores[most_similar_chunk_index]

    return chunks[most_similar_chunk_index], highest_score

while True:
    context = extractedInformation
    question = input("Ask your question (or 'q' to exit): ")

    if question.lower() == 'q':
        print("Exiting question answering session.")
        break

    chunk_size_limit = 264
    chunks = [context[i:i + chunk_size_limit] for i in range(0, len(context), chunk_size_limit)]
    most_similar_chunk, highest_score = get_most_similar_chunk(question, chunks)

    if highest_score == 0:
        print("No answer found in the provided context.")
    else:
        answer = answer_question(question, most_similar_chunk)
        if answer:
            print("Answer:", answer)
        else:
            print("No answer found in the provided context.")
```
